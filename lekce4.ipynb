```python
pocet_obyvatel = pd.read_sql("pocet_obyvatel", con=engine)
pocet_obyvatel
```

```python
pocet_bytu = pd.read_sql("pocet_bytu", con=engine)
pocet_bytu
```

```python
spojene = pocet_obyvatel.merge(pocet_bytu, on='obec')
spojene
```

```python
nad1000 = spojene.loc[spojene['pocet_obyvatel'] > 1000]
nad1000["na 100 oby"] = nad1000["pocet_bytu"] / nad1000["pocet_obyvatel"] * 100

nad1000 = nad1000.sort_values(by=["na 100 oby"], ascending=False).head(5)
nad1000
```

```python
datajoin = pd.read_sql("SELECT pocet_obyvatel.obec, pocet_obyvatel.pocet_obyvatel, pocet_bytu.pocet_bytu FROM pocet_obyvatel INNER JOIN pocet_bytu ON pocet_obyvatel.obec=pocet_bytu.obec WHERE pocet_obyvatel.pocet_obyvatel > 1000 ORDER BY pocet_bytu.pocet_bytu desc", con=engine)
datajoin
```
