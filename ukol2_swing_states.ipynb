```python
import requests
import pandas as pd
import numpy as np

```

```python
with requests.get("https://raw.githubusercontent.com/pesikj/progr2-python/master/data/1976-2020-president.csv") as r:
  open("1976-2020-president.csv", 'w', encoding="utf-8").write(r.text)

volby_data = pd.read_csv("1976-2020-president.csv")
volby_data.head()
```

```python
#pořadí kandidátů podle roku a státu
volby_data["Rank"] = volby_data.groupby(["year","state"])["candidatevotes"].rank(ascending=False)
volby_data.head()
```

```python
#odstranění kandidátů, kteří nevyhráli
volby_vitezove = volby_data.drop(volby_data[volby_data.Rank > 1].index)
volby_vitezove.head()
```

```python
#vytvoření sloupce s porovnáním s předchozími volbami
volby_vitezove = volby_vitezove.sort_values(["state","year"])
volby_vitezove["winner prev.year"] = volby_vitezove.groupby(["state"])["party_detailed"].shift()
volby_vitezove
```

```python
#vytvoření sloupce, kde je poznat, jestli došlo ke změně
volby_vitezove["diff.winner"]= np.where(volby_vitezove["party_detailed"] != volby_vitezove["winner prev.year"],1,0)
#odstranění řádků, kde není srovnání s předchozím vítězem
volby_vitezove.dropna(subset=["winner prev.year"], inplace=True)
volby_vitezove
```

```python
pivot_volby = pd.pivot_table(volby_vitezove, index="state",values="diff.winner", aggfunc="sum")
pivot_volby = pivot_volby.sort_values(by=["diff.winner"], ascending=False)
pivot_volby
```
