```python
import requests

r = requests.get("https://raw.githubusercontent.com/pesikj/progr2-python/master/data/Fish.csv")
with open("Fish.csv", "wb") as f:
  f.write(r.content)
```

```python
import pandas as pd 
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv("Fish.csv")
data
```

```python
mod = smf.ols(formula="Weight ~ Length2", data=data)
res = mod.fit()
res.summary()
```

Model je poměrně přesný, popisuje 84% dat

```python
mod2 = smf.ols(formula="Weight ~ Length2 + Height", data=data)
res2 = mod2.fit()
res2.summary()
```

kvalita modelu se mírně zlepšila

```python
#target encoding
druhy = data.groupby('Species')['Weight'].mean().sort_values()
data["druhy_prumerna_vaha"] = data["Species"].map(druhy)
data

```

```python
mod3 = smf.ols(formula="Weight ~ Length2 + Height + druhy_prumerna_vaha", data=data)
res3 = mod3.fit()
res3.summary()
```

```python
vypocet = pd.DataFrame({"Length2": [15.2, 40],
                         "Height": [3.5, 4],
                         "druhy_prumerna_vaha": [11.17, 829]})
res3.predict(vypocet)
```
