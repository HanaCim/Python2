```python
import pandas as pd
import requests
```

```python
r = requests.get("https://raw.githubusercontent.com/lutydlitatova/czechitas-datasets/main/datasets/lexikon-zvirat.csv")
open("lexikon-zvirat.csv", "wb").write(r.content)
```

```python
data = pd.read_csv("lexikon-zvirat.csv", delimiter=";", index_col='id')
data.dropna(axis=0, how='all', inplace=True)
data.dropna(axis=1, how='all', inplace=True)
data.head()
```

```python
#vytvoření funkce pro tvorbu popisku
def popisek (radek):
  return f"{radek.title} se živí především: {radek.food}, konkrétně pak {radek.food_note}.\n Jak toto zvíře poznáme? {radek.description}"
```

```python
#uplatnění funkce pro dataframe
data["popisek"] = data.apply(popisek, axis=1)
data.head()
```
