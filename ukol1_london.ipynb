```python
import requests
import pandas as pd
import numpy as np
```

```python
r = requests.get("https://raw.githubusercontent.com/pesikj/progr2-python/master/data/london_merged.csv")
open("london_merged.csv", 'wb').write(r.content)

data2 = pd.read_csv("london_merged.csv")

data2 
```

```python
import datetime
data2["timestamp"] = pd.to_datetime(data2["timestamp"])
data2["year"] = data2["timestamp"].dt.year
data2["count"] = 1
data2
```

```python
london_pivot1 = pd.pivot_table(data2, index="weather_code", columns="year", values="count", aggfunc="sum")
london_pivot1
```
