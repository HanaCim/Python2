```python
import requests
import pandas as pd
import seaborn as sns
import statistics

```

```python

r = requests.get("https://raw.githubusercontent.com/pesikj/progr2-python/master/data/crypto_prices.csv")
open("crypto_prices.csv", "wb").write(r.content)
```

```python
data = pd.read_csv("crypto_prices.csv")
data = data[["Symbol", "Date", "Close"]]
data.sort_values(by=["Symbol","Date"],inplace=True)
data
```

```python
data["zmena"] = data.groupby("Symbol")["Close"].pct_change()
data
```

```python
DOGE = data.loc[data['Symbol'] == 'DOGE']
DOGE = DOGE.drop(["Symbol","Close"], axis=1)
DOGE['zmena'] = DOGE['zmena'] + 1

zmena = DOGE['zmena'].to_list()
zmena_bez_nan = [x for x in zmena if str(x) != 'nan']
zmena_bez_nan
```

```python
vysledek = (statistics.geometric_mean(zmena_bez_nan) - 1)
vysledek
```
