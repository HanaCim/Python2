```python
import requests
import pandas as pd 

with requests.get("https://raw.githubusercontent.com/pesikj/progr2-python/master/data/air_polution_ukol.csv") as r:
  open("air_polution_ukol.csv", 'w', encoding="utf-8").write(r.text)
```

```python
data = pd.read_csv("air_polution_ukol.csv")
data["date"] = pd.to_datetime(data["date"])
data["year"] = data["date"].dt.year
data["month"] = data["date"].dt.month
data
```

```python
years= [2019, 2020]

leden = data[(data['month'] == 1) & data['year'].isin(years)]
leden
```

H0 - počty částic se rovnají
H1 - počty částic jsou různé

```python
from scipy.stats import mannwhitneyu

x = leden[leden["year"] == 2019]["pm25"]
y = leden[leden["year"] == 2020]["pm25"]
print(x.shape)
print(y.shape)
mannwhitneyu(x, y)
```

P-hodnota je menší než naše hladina významnosti 0,05, proto můžeme zamítnout hypotézu, že počet částic byl ve srovnávaných letech stejný.
