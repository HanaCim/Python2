```python
import requests
import pandas as pd
import numpy
```

```python
r = requests.get("https://raw.githubusercontent.com/pesikj/progr2-python/master/data/baroko_half_marathon.csv")
open("baroko_half_marathon.csv", 'wb').write(r.content)

data = pd.read_csv("baroko_half_marathon.csv")
data
```

```python
data = data.sort_values(["Jméno závodníka", "Ročník", "Rok závodu"])
data
```

```python
data["FINISH"] = pd.to_datetime(data["FINISH"])
data["Prev.Year"] = data.groupby(["Jméno závodníka"])["FINISH"].shift()
data.head()
```

```python
data.dropna(subset=['Prev.Year'], inplace=True)
data
```

```python
data["rozdil"] = data["FINISH"] - data["Prev.Year"]
data
```

```python
data["lepsi_horsi"] = numpy.where(data["rozdil"] < pd.Timedelta("P0D"), "Lepší", "Horší")
data
```
